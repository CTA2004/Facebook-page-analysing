## Data crawling

```python
from facebook_scraper import get_posts
import pandas as pd
import numpy as np
```

```python
FANPAGE_LINK = "thaythuantoan"
FOLDER_PATH = "c:/Users/HP/github-classroom/IAI-UET/final-project-CTA2004"
COOKIE_PATH = "C:/Users/HP/github-classroom/IAI-UET/final-project-CTA2004/www.facebook.com_cookies.txt"

PAGES_NUMBER = 10
```

```python
post_list = []
for post in get_posts(FANPAGE_LINK,
                    options={"comments": True, "reactions": True, "allow_extra_requests": True},
                    extra_info=True, pages=PAGES_NUMBER, cookies=COOKIE_PATH):
    print(post)
    post_list.append(post)
```

```python
raw_df = pd.DataFrame(columns=post_list[0].keys(), index=range(len(post_list)), data=post_list)

path=FOLDER_PATH + FANPAGE_LINK + ".csv"
raw_df.to_csv(path, index=False)
print(path)
```

 ## Data pre_processing

```python
new_df = raw_df.dropna(axis = 1, how='any')
```

```python
new_df = new_df.drop("was_live", axis= 1)
new_df = new_df.drop("is_live", axis= 1)
```

```python
new_df['reactions'] = new_df['reactions'].apply(lambda x : dict(eval(x)))
new_df_reactions = new_df['reactions'].apply(pd.Series)
```

```python
new_df = pd.concat([new_df, new_df_reactions], axis=1)
```

```python
new_df = new_df.fillna(value = 0)
```

```python
new_df['time'] = pd.to_datetime(new_df['time'])
new_df['date'] = new_df['time'].dt.date
new_df['hour'] = new_df['time'].dt.hour
```

```python
import re

def loai_bo_ky_tu_dac_biet(post_text):
    post_text_loai_bo_dac_biet = re.sub(r'[^a-zA-Z0-9\s]', '', post_text)
    
    return post_text_loai_bo_dac_biet

for post_text in new_df["post_text"]:
    loai_bo_ky_tu_dac_biet(post_text)
```
